```python
#Importe Numpy y Pandas de acuerdo con las convenciones explicadas.
import numpy as np
import pandas as pd
```

```python
#Lea el archivo taller7.npy, y asignelo a una variable llamada datos.

datos = np.load("/content/drive/MyDrive/Universidad/Semestre9/Computación científico y analítica/Corte 2/Semana7.npy")
print(datos)
```

```python
#Cuántos elementos tiene el arreglo? Utilice funciones de numpy para responder a esta pregunta.
cantidad = len(datos)
print("Cantidad de datos en el arreglo: ",cantidad)
```

```python
#¿Cuál es el valor promedio del arreglo? Utilice funciones de numpy para responder a esta pregunta.
promedio = np.mean(datos)
print("El promedio de datos del arreglo es: ",promedio)
```

```python
#¿Cuál es la suma total de todos los elementos del arreglo? Utilice funciones de numpy para responder a esta pregunta.
suma = np.sum(datos)
print("La suma de todos los datos del arreglo es: ",suma)
```

```python
#¿Cuál es el valor máximo del arreglo? Utilice funciones de numpy para responder a esta pregunta.
maximo = np.max(datos)
print("El valor maximo de todos los datos del arreglo es: ",maximo)
```

```python
#¿Cuál es el valor mínimo del arreglo? Utilice funciones de numpy para responder a esta pregunta.

minimo = np.min(datos)
print("El valor minimo de todos los datos del arreglo es: ",minimo)
```

```python
#Cree una Serie de Pandas a partir del arreglo datos.
serie = pd.Series(datos)
print("Serie: ",serie)
```

```python
#Muestre los primeros 7 elementos de la Serie. Utilice funcionalidades de Pandas.
serie.head(n=7)
```

```python
#Muestre los últimos 4 elementos de la Serie. Utilice funcionalidades de Pandas.
serie.tail(n=4)
```

```python
#¿Cuáles son los índices de esta Serie? ¿Por qué?
serie.iloc[[0,99]]
```

####La serie considerando que esta tiene valores indices por defecto, por el cual si tomamos el primero y el ultimo indice de la serie, se visualiza que esta constituida de un valor minímo de 0 y un valor maxímo de 99, en los indices de la serie construida.

```python
#,Con una sola instrucción, presente los valores en las posiciones 33, 27, 90, 7, y 14, en ese orden.
serie.iloc[[33, 27, 90, 7 ,14]]
```

```python
#Con una sola instrucción, presente los valores cuyos índices son 33, 27, 90, 7, y 14, en ese orden.
serie.loc[[33, 27, 90, 7 ,14]]
```

####¿Por qué son iguales los dos resultados anteriores? Elabore y explique su respuesta.

####La explicación de esta se refiere a que por defecto lanza la lista ordenada independientemente de que este sacando el valor del arreglo, es decir, si yo consulto por el valor de arreglo, lanzará un indice (por defecto), por otro lado, si consultamos a través de los indices, estos estan definidos de 0 a n, es por esto que al momento de consultar nos retorna la misma información.



```python
#Con una sola instrucción, presente los valores de la última posición y la primera posición de la Serie.
serie.iloc[[0,99]]
```

```python
#Cree una copia del arreglo datos, y cambie su geometría a 50 filas y 2 columnas. 

copia = datos.flatten().reshape(50,2)

print(copia)
```

```python
#Cree un DataFrame a partir del arreglo bidimensional que creo en el punto anterio
df = pd.DataFrame(copia)
df
```

```python
#Con una sola instrucción, presente los valores de ambas columnas en las posiciones 5, 10, 15, 20, y 25, en ese orden.
df.iloc[[5, 10, 15, 20,  25]]
```

```python
#Con una sola instrucción, presente los valores de la segunda columna en las posiciones 5, 10, 15, 20, y 25, en ese orden.
df[1].iloc[[5, 10, 15, 20,  25]]
```

```python
#Con una sola instrucción, presente los valores de la primera columna cuyos índices son 5, 10, 15, 20, y 25, en ese orden.
df[0].iloc[[5, 10, 15, 20,  25]]
```
