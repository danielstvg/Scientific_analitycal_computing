```python
#Para todas las gráficas, asegúrese de proveer un título adecuado, nombres para los ejes, y una leyenda de ser necesario. 
#Si una misma gráfica incluye dos series distintas, asegúrese de utilizar colores, líneas y/o marcadores distintos para diferenciarlas:

#Grafique los datos de Espesor y de Disipación en una misma gráfica.
plt.figure(figsize=(20,15))
data["Discipacion"].plot(kind='area', color="blue", alpha=0.6, figsize=(20, 10))
data["Espesor"].plot(kind='area', color="orangered",alpha=0.7, figsize=(20, 10))
plt.plot(data["Espesor"],  color="red", linewidth=2.5, linestyle="--",  marker="")
plt.plot(data["Discipacion"],  color="blue",  linewidth=2.5, linestyle=":",  marker="")
plt.title("Grafico de comportamiento de la discipación con respecto al espesor")
plt.xlabel("Registros")
plt.ylabel("Discipación y Espesor")
plt.legend()
```

```python
#Muestre los siguientes histogramas como dos subgráficas de una misma figura

fig, (ax0, ax1) = plt.subplots(figsize=(20,8),ncols=2)

x = data["Espesor"]
y = data["Discipacion"]

ax0.hist(x, label="Espesor", bins=40, color="orangered")
ax0.title.set_text("Espesor")
ax0.set_xlabel('Registros') #nombre del eje x
ax0.set_ylabel('Nivel de espesor') #nombre del eje y
ax1.hist(y, label="Discipacion", bins=40, color="blue")
ax1.title.set_text("Discipacion")
ax1.set_xlabel('Registros') #nombre del eje x
ax1.set_ylabel('Nivel de discipacion') #nombre del eje y
plt.tight_layout()
plt.show()
```

 * El comportamiento con respecto a la gráfica de area, presenta la distribución de los datos de caracter como si fuera una linea de tendencia lineal o lo más semejante a una linea recta.  Este tipo de gráfica presenta la información de forma que sombrea el área de los datos presentados que determina los rangos acumulados a lo largo de los registros que se encuentran en el dataset.

 * El comportamiento en la gráfica del histograma tiene una distribución más volatíl, estimando la información de los datos correspondientes a este tipo de gráficas presenta con cada una de las barras es una representación que es proporcional a la frecuencia de los datos en el dataset, por lo cual, esto indica que tiene un mayor grado de frecuencia entre los valores 1.4 y 1.6. Este histograma permite visualizar la distribución de los datos de caracter de una visión en forma de grupos, relacionandolos con una frecuencia o tendencia a nivel de los datos presentados sobre el espesor.

```python
# Muestre el correspondiente diagrama de dispersión, y describa sus observaciones.


fig, (ax0, ax1) = plt.subplots(figsize=(20,8),ncols=2)

x = data["Espesor"]
y = data["Discipacion"]

ax0.scatter(x,y, c='red', edgecolors='w')
ax0.title.set_text("Espesor vs Discipación")
ax0.set_xlabel('Nivel de discipación') #nombre del eje x
ax0.set_ylabel('Nivel de espesor') #nombre del eje y
ax1.scatter(y,x, c='blue', edgecolors='w')
ax1.title.set_text("Discipacion vs Espesor")
ax1.set_xlabel('Nivel de espesor') #nombre del eje x
ax1.set_ylabel('Nivel de discipación') #nombre del eje y
plt.tight_layout()
plt.show()
```

## Que relación existe entre los datos de Espesor y de Discipación?

* En relación entre Espesor con respecto a discipación tiene un comportamiento casi similar al que se presento a la gráfica del área, por lo cual, no cambia demasiado, siendo un comportamiento lineal, pero en sus ultimos registros con un comportamiento decreciente

* En relacioón entre Discipación con respecto a Espsor este tiene un comportamiento lineal, creciente y los puntos que toma sobre los ejex X y Y, parecia devolverse, sin embargo se entiende que es porque la gráfica esta sobre valores invertidos.
