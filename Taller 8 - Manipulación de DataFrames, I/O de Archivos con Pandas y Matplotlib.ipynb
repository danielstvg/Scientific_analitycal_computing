```python
#2. Importe Numpy, Pandas y Matplotlib de acuerdo con las convenciones explicadas.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#3. Lea el archivo BOG_Temp_SWR_CloudCover_2020_0922-1006_hourly.csv, y asignelo a un DataFrame llamado data. Asegúrese que la columna "Fecha" 
#sea tomada como el índice para las filas del dataframe.


df = pd.read_csv("/content/drive/MyDrive/Universidad/Semestre9/Computación científico y analítica/Corte 2/BOG_Temp_SWR_CloudCover_2020_0922-1006_hourly.csv", index_col="Fecha")
df.head()
```

```python
#6. Asegúrese que la columna "Radiacion" sea de tipo np.float32.
df['Radiacion'].dtype
```

```python
import string

#7. Adicione una columna al DataFrame, llamada "Letras", con letras mayúsculas entre la A y la O.
for i in range(15):
  df['Letras'] = [chr(i) for i in range(ord('A'),ord('O')+1)]
df
```

```python
#8. Adicione una columna al DataFrame, llamada "Numeros", con números entre el 0 y el 14, entre la columna "Temperatura" y la columna "Cobertura".
df.insert(2,"Numeros", [int(i) for i in range(0,15)])
df
```

```python
#9. Cambie permanentemente el nombre de la columna "Numeros" por "Col1".
df = df.rename(columns={'Numeros':'Col1'})
df
```

```python
#10. Cambie permanentemente el nombre de la columna "Letras" por "Col0".
df = df.rename(columns={'Letras':'Col0'})
df
```

```python
#11. Borre permanentemente la columna "Col0" utilizando el método .drop().
df = df.drop("Col0", axis=1)
df
```

```python
#12. Borre permanentemente la columna "Col1" utilizando el método .pop().
df.pop("Col1")
df
```

```python
#13. Grafique los datos de Temperatura. Asegúrese de proveer un título adecuado, nombres para los ejes, y una leyenda si lo considera necesario.
plt.figure(figsize=(17,8))
plt.plot(df["Temperatura"], label = "Temperatura")
plt.title("Grafico de temperatura")
plt.xlabel("Fecha")
plt.ylabel("Temperatura")
plt.legend()
```

```python
#14. Grafique los datos de Cobertura. Asegúrese de proveer un título adecuado, nombres para los ejes, y una leyenda si lo considera necesario.
plt.figure(figsize=(17,8))
plt.plot(df["Cobertura"], label = "Cobertura")
plt.title("Grafico de Cobertura")
plt.xlabel("Fecha")
plt.ylabel("Cobertura")
plt.legend()
```

```python
#15. Grafique los datos de Radiacion. Asegúrese de proveer un título adecuado, nombres para los ejes, y una leyenda si lo considera necesario.
plt.figure(figsize=(17,8))
plt.plot(df["Radiacion"], label = "Radiacion")
plt.title("Grafico de Radiacion")
plt.xlabel("Fecha")
plt.ylabel("Radiacion")
plt.legend()
```

```python
#16. Cree una copia del dataframe llamada norm_data con el siguiente comando:
#16. norm_data = ( data - data.min()) / (data.max() - data.min())
#Esto creará una copia normalizada, con valores entre 0 y 1 para todas las columnas, lo que le permitirá responder las siguientes preguntas.

df2 = ( df - df.min()) / (df.max() - df.min())
df2
```

```python
#17. Grafique los datos normalizados de Temperatura y Radiación en una sola gráfica. 
#Asegúrese de proveer un título adecuado, nombres para los ejes, y una leyenda si lo considera necesario.
plt.figure(figsize=(17,8))
plt.plot(df2["Temperatura"],  color="blue", linewidth=2.5, linestyle="-", label="Temperatura", marker="*")
plt.plot(df2["Radiacion"],  color="red",  linewidth=2.5, linestyle="-", label="Radiación", marker="h")
plt.title("Grafico de Radiacion")
plt.xlabel("Fecha")
plt.ylabel("Radiacion")
plt.legend()
```

#### La relación entre los valores del dataframe, se presenta de forma similar, la información no es tan dispersa con respecto a la normalización de los datos previamente aplicado, a partir del tiempo este refleja un comportamiento volátil

```python
#18. Grafique los datos normalizados de Cobertura y Radiación en una sola gráfica.
# Asegúrese de proveer un título adecuado, nombres para los ejes, y una leyenda si lo considera necesario

plt.figure(figsize=(17,8))
plt.plot(df2["Cobertura"],  color="blue", linewidth=2.5, linestyle="-", label="Cobertura")
plt.plot(df2["Radiacion"],  color="red",  linewidth=2.5, linestyle="-", label="Radiación")
plt.title("Grafico de Radiacion")
plt.xlabel("Fecha")
plt.ylabel("Radiacion")
plt.legend()
```

#### La relación entre la cobertura y la radición, aunque en un principio tiene un comportamiento opuesto, con respecto a la cobertura total de nubes en porcentaje, y la radiación en watts en un periodo de tiempo de 25 días. este comportamiento refleja una relación volátil.
