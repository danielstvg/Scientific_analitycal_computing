5. Calcule una nueva columa llamada "Caja", en donde para cada día calcule la cantidad de dinero que la tienda recaudó por la venta del producto.

```python
for i in range(len(df)):
  df['Caja'] = [float(i) for i in df['Precios']*df['Ventas']]
df
```

6. Calcule una nueva columna llamada "Cartera", en donde para cada día calcule la cantidad de dinero que la tienda tuvo que devolver a causa de las devoluciones de producto.

```python
for i in range(len(df)):
  df['Cartera'] = [float(i) for i in df['Devoluciones']*df['Precios']]
df.head(30)
```

7. Calcule una nueva columna llamada "Neto", en donde para cada día se calcule el total neto de restarle el dinero de las devoluciones al total de ventas.

```python
for i in range(len(df)):
  df['Neto'] = [float(i) for i in df['Caja']-df['Cartera']]
df
```

14. Grafique el número promedio de ventas para cada mes del año 2019 presente en el DatFrame (Downsampling). Describa y analice sus resultados.

```python
plt.figure(figsize=(10,8))
plt.title("Promedio de ventas del año 2019 (Downsampling)")
plt.xlabel("Fecha")
plt.ylabel("Valor promedio")
downsampling = df['Ventas'].resample('W').mean()
downsampling.head(20)
downsampling['2019'].plot()
```

**RTA 14**
Se representan los valores de ventas en las semanas del segundo semestre del 2019, estimando una variación de los datos de forma contraria, se observan picos demasiado pronunciados, lo que indica que en las semanas son muy volatiles, por medio de este metodo (DownSampling) se reduce el numero total de la serie

15. Grafique el número promedio de ventas para cada semana del cuarto trimestre del año 2019 presente en el DatFrame (Upsampling). Describa y analice sus resultados.

```python
plt.figure(figsize=(10,8))
plt.title("Promedio de ventas del mes de Octubre a Diciembre del 2019 (Upsampling)")
plt.xlabel("Fecha")
plt.ylabel("Valor promedio")
downsampling = df['Ventas'].resample('W').mean().interpolate()
downsampling.head(20)
downsampling['2019-10':'2019-12'].plot()
```

20. Calcule una nueva columna llamada "Diferencia", donde para cada día calcule la diferencia en precio de venta del producto, con respecto al día anterior. Para el primer registro del DataFrame este valor debe inicializarse como cero.
