```python
# 12. Revierta los arreglos A y B a sus dimensiones originales de forma permanente, usando un solo comando, y multiplique cada uno de sus elementos 
#por la suma de todos los elementos del arreglo B. 
#Asegúrese de utilizar una implementación eficiente de dichas operaciones, y no utilice bucles. Guarde este resultado como la variable X.

A=A.reshape(5,8)
B=B.reshape(8,5)
print("A: \n",A)
print("B: \n",B)

X = A*np.sum(B)

print("\nX: \n",X)


```
