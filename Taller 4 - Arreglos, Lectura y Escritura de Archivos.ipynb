```
#Cargue los contenidos del archivo en un objeto de python. Asegúrese de remover los saltos de línea ('\n') que puedan haberse incluido en la lectura
import pandas as pd
import numpy as np


a = np.fromfile('/content/drive/MyDrive/Universidad/Semestre9/taller4.dat', dtype=np.int8, sep=',')
print(a)
```

```
#Convierta dicho objeto con sus datos en una lista de python, y luego conviértala en un Arreglo de Numpy. Asegúrese de que el tipo de dato de su arreglo sea 'float32'.

b = np.array(a,dtype='float32')
print(b)
```

```
#Genere una vista con 12 objetos contiguos a lo largo del arreglo
vista = b[:12]
print(vista)
```

```
#Genere una copia del arreglo original en una variable nueva.

copia = np.array(b)
print(copia)
```

```
#Genere una vista de dicha copia como una matriz de 5 filas y 10 columnas, y asígnela a la variable X
 
X =copia.reshape(5,10)
print(X)
```

```
#Presente la  matriz X de 5x10, pero transpuesta.
X.transpose()
```

```
#Genere una vista de los datos originales como una matriz de 5 filas y 10 columnas, y asígnela a la variable Y

Y = b.reshape(5,10)
print(Y)
```

```
#Cambie la primera fila de X, por ceros.

X[:1] = 0
X
```

```
#Realice una comparación de igualdad entre X y Y, y guarde la matriz Booleana resultante como la variable COMP

COMP = (X==Y)
print(COMP)
```

```
#Realice una comparación X < Y, y guarde la matriz Booleana resultante como la variable COMP2

COMP2 = X < Y
print(COMP2)
```

```
#imprima los resultados de las siguiente instrucciones, y explique por qué obtiene cada resultado:
COMP.all()
```

####El reultado es la comparación de cada uno de los elementos entre si de la matriz, es decir si este cumple en aspectos lógicos de verdad, en caso tal de que toda la matriz sea igual se obtendra un resultado afirmativo (True), en caso contrario si la matriz no es igual en su totalidad, se obtendra un resultado negativo (Falso)

```
#imprima los resultados de las siguiente instrucciones, y explique por qué obtiene cada resultado:
COMP.any()
```

####El resultado es la comparación de cada uno de los elementos entre si de la matriz, imponiendo la regla de si al menos uno de los elementos se parece a cuaquier elemento interno de la matriz. 

```
#Realice la operación lógica AND entre la matriz COMP negada y la matriz COMP2; guarde su resultante matriz Booleana como la variable ResAND.
ResAND = ~COMP & COMP2
print(ResAND)
```

```
#Aplique la matriz ResAND como máscara sobre la matriz X, e imprima su resultado. Explique por qué obtuvo un vector de 9 ceros.
X[ResAND]
```

El vector resultandte que se evidencia en la mascara del resultado de ResAnd se traslapa con los resultados de la matriz X, al ejecutarlo solo aquellos unicos valores que tengan un valor de verdad afirmativo se imprimen sobre el tablero
